# Visualize results for runtime experiments

```python
# necessary imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plot differences in FairDen with fixed vs adaptive minpts parameter
def plot_dim2_experiment():
    files_list = ['dim_experiment.csv','dim_experiment2.csv']
    colors = ['#FF1F5B','#004D40']
    exp_list = []
    my_path = 'results/'
    for file in files_list:
        experiment = pd.read_csv(my_path+file)
        if file =='dim_experiment.csv':
            experiment['Algorithm'] = ['FairDen (adjusted)']*len(experiment)
            experiment['Dim'] = experiment['dim']
        if file =='dim_experiment2.csv':
            experiment['Algorithm'] = ['FairDen (fixed)']*len(experiment)
        exp_list.append(experiment)
    experiment = pd.concat(exp_list, ignore_index=True)
    experiment = experiment[experiment['Dim']!=1000]
    sns.set(font_scale=1.5)
    sns.set_style("whitegrid")
    sns.lineplot(data=experiment, x="Dim", y="Time", hue="Algorithm",style="Algorithm",palette = colors)
    plt.savefig(my_path+'dim_fairden.svg')
plot_dim2_experiment()
```

```python
# plot experiment with varying dataset dimensions
def plot_dim_experiment():
    files_list = ['dim_experiment2.csv','o_dim2_experiment.csv', 'fairlet_dim2_experiment.csv']
    colors = ['#FF1F5B','#ffc61e','#009ade','#f28522','#af58ba']
    colors_wo_fairlet = ['#FF1F5B','#ffc61e','#009ade','#af58ba']
    exp_list = []
    my_path = 'results/'
    for file in files_list:
        experiment = pd.read_csv(my_path+file)
        if file =='dim_experiment.csv':
            experiment['Algorithm'] = ['FairDen (adjusted)']*len(experiment)
            experiment['Dim'] = experiment['dim']
        if file =='dim_experiment2.csv':
            experiment['Algorithm'] = experiment['algorithm']
        exp_list.append(experiment)
    experiment = pd.concat(exp_list, ignore_index=True)
    experiment.sort_values(by=['Algorithm'], ascending=True, inplace=True)
    experiment1 = experiment[experiment['Algorithm']!='Fairlet']
    sns.set(font_scale=1.5)
    sns.set_style("whitegrid")
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    sns.lineplot(data=experiment, x="Dim", y="Time", hue="Algorithm",style="Algorithm",palette = colors, ax = axes[0],dashes= True,legend=False)
    sns.lineplot(data=experiment1, x="Dim", y="Time", hue="Algorithm",style="Algorithm",palette = colors_wo_fairlet,ax = axes[1],dashes= True,legend=False)
    plt.savefig(my_path+'dims_experiment.svg')
plot_dim_experiment()
```

```python
# plot experiment with varying numbers of datapoints
def plot_n_experiment():
    files_list = ['n_experiment.csv','o_n_experiment.csv', 'fairlet_n_experiment.csv']
    colors = ['#FF1F5B','#ffc61e','#009ade','#f28522','#af58ba']
    exp_list = []
    my_path = 'results/'
    for file in files_list:
        experiment = pd.read_csv(my_path+file)
        exp_list.append(experiment)
    experiment = pd.concat(exp_list, ignore_index=True)
    experiment = experiment[experiment['n']!=20000]
    experiment.sort_values(by=['Algorithm'], ascending=True, inplace=True)
    sns.set(font_scale=1.5)
    sns.set_style("whitegrid")
    fig, axes = plt.subplots(1,1)
    sns.lineplot(data=experiment, x="n", y="Time", hue="Algorithm",style="Algorithm",palette = colors,ax = axes,dashes= True,legend=False)
    plt.savefig(my_path+'n_experiment.svg')
plot_n_experiment()
```
